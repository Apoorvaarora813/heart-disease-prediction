## which one to use? pypplot vs matplotlib OO method 
* When plotting something quickly, okay to use pyplot method 
* When plotting something advance use OO method 

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# pyplot method 
over_50.plot(kind='scatter',x='age',y='chol',c='target');
```

```python
# OO method using pyplot
fig, ax =plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);
ax.set_xlim([45,100])
```

```python
## oo method from scratch 
fig,ax=plt.subplots(figsize=(10,6))
#plot the data 
scatter =ax.scatter(x=over_50['age'],
                   y=over_50['chol'],
                   c=over_50['target']);

#customize the plot
ax.set(title="heart disease and cholestrol levels",
       xlabel="age",
       ylabel="cholestrol");
# add a legend 
ax.legend(*scatter.legend_elements(),title="target");

# add a horizontal line 
ax.axhline(over_50["chol"].mean(), linestyle='--');
```
